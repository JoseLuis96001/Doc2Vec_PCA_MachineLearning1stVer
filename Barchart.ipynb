```python
import plotly.graph_objects as go
animals=['SGD', 'SVMs', 'Random Forest','AdaBoost','KNeighbors']
y1=[0.84, 0.86, 0.85, 0.83, 0.85]
y2=[0.78, 0.82, 0.81, 0.80, 0.80]

fig = go.Figure(data=[
    go.Bar(name='Accuracy', x=animals, y=y1,text=y1,textposition='auto'),
    go.Bar(name='Accuracy after PCA', x=animals, y=y2,text=y2,textposition='auto')
])
# Change the bar mode
fig.update_layout(barmode='group',
                  autosize=False,
                  width=800,
                  height=400,)
fig.update_layout(legend=dict(
    orientation="h",
    yanchor="bottom",
    y=1.02,
    xanchor="right",
    x=1
),
font=dict(
        family="Droid Serif",
        size=21,
    ))



fig.show()
```
